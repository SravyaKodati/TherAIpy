```python
import gradio as gr
import json
import faiss
import torch
from unsloth import FastLanguageModel
from sentence_transformers import SentenceTransformer
```

```python
faiss_index = faiss.read_index("faiss_index")
print("faiss index loaded successfully!")

with open("metadata.json", "r") as f:
    metadata = json.load(f)
```

```python
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

max_seq_length = 10000
dtype = None
load_in_4bit = True

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="SravyaKodati/TherAIpy-LoRA",
    max_seq_length= max_seq_length,
    dtype= dtype,
    load_in_4bit= load_in_4bit
)

FastLanguageModel.for_inference(model)
```

```python
def search(query, tokenizer, embedding_model, faiss_index, metadata_file, top_k=5):
    """
    Searches for the most relevant chunks given a query.

    Args:
        query (str): The search query.
        tokenizer: Pre-initialized tokenizer.
        embedding_model: The embedding model used for encoding.
        faiss_index: Preloaded FAISS index for search.
        metadata_file (str): Path to the metadata file for retrieving chunk information.
        top_k (int): Number of top results to return.

    Returns:
        list: A list of top-k results with metadata (text, page numbers, etc.).
    """

    print("Query:", query)
    query_embedding = embedding_model.encode([query])  # SentenceTransformer expects a list of strings
    print("Query embedding shape:", query_embedding.shape)

    distances, indices = faiss_index.search(query_embedding, top_k)

    with open(metadata_file, "r") as f:
        metadata = json.load(f)

    results = []
    for idx, distance in zip(indices[0], distances[0]):
        if idx == -1:
            continue
        result = metadata.get(str(idx), {})
        result["distance"] = distance
        results.append(result)

    return results


def retrieve(query, results):
  top_results = [r["text"] for r in results]
  top_result= top_results[0]

  return top_result
```

```python
def generate_response(query):
  results = search(query, tokenizer, embedding_model, faiss_index, metadata_file="metadata.json", top_k=5)
  rag_context = retrieve(query, results)

  alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

  ### Instruction:
  {}

  ### Input:
  {}

  ### Response:
  {}"""

  instruction = """You're a mental health chatbot providing thoughtful, helpful, and safe responses to user queries.
  Your answers should be empathetic but realistic, focusing on providing actionable advice without being overly optimistic.
  You're going to get two inputs. One is the query from the patients, and another one is a RAG context retrieved from
  some mental health books. RAG context is supposed to help you frame answers. But sometimes, it might give you irrelavant information
  which is unrelated to the query. If that's the case, please ignore the RAG context and give appropriate answer to the query.
  In either answer the questions like a therapist."""

  prompt = f"Query: {query} RAG context: {rag_context}"
  inputs = tokenizer(
                      [
                      alpaca_prompt.format(
                      instruction,
                      prompt,
                        "",
                      )
                      ], return_tensors = "pt").to("cuda")

  outputs = model.generate(**inputs, max_new_tokens=1000, use_cache=True)
  response = tokenizer.batch_decode(outputs, skip_special_tokens=True)
  clean_response= response[0].split("<|end_of_text|>")[0].split("### Response:\n")[-1].strip()

  return clean_response
```

# Testing

```python
import pandas as pd
import csv

def process_patient_questions(input_csv, output_csv):
    df = pd.read_csv(input_csv)
    df['Chatbot Response'] = ''

    for index, row in df.iterrows():
        question = row['Patient Question']
        response = generate_response(question)
        df.at[index, 'Chatbot Response'] = response
        print(f"Processed question {index + 1} of {len(df)}")

    df.to_csv(output_csv, index=False, quoting=csv.QUOTE_ALL)
    print(f"Processing complete. Results saved to {output_csv}")
```

```python
input_csv = 'test questions.csv'
output_csv = 'test questions with responses.csv'

process_patient_questions(input_csv, output_csv)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

def generate_embeddings(chunks, model_name="all-MiniLM-L6-v2"):
    """
    Generates embeddings for text chunks using a SentenceTransformer model.
    Args:
        chunks (list): List of text chunks.
        model_name (str): Name of the embedding model.
    Returns:
        list: A list of embeddings.
    """
    model = SentenceTransformer(model_name)
    embeddings = model.encode(chunks, show_progress_bar=True)
    return embeddings

def calculate_cosine_similarity(input_csv, output_csv):
    df = pd.read_csv(input_csv)

    if 'Therapist Answer' not in df.columns or 'Chatbot Response' not in df.columns:
        raise ValueError("Input CSV must contain 'Therapist Answer' and 'Chatbot Response' columns.")

    therapist_answers = df['Therapist Answer'].tolist()
    chatbot_responses = df['Chatbot Response'].tolist()

    therapist_embeddings = generate_embeddings(therapist_answers)
    chatbot_embeddings = generate_embeddings(chatbot_responses)

    similarities = cosine_similarity(therapist_embeddings, chatbot_embeddings)

    df['Cosine Similarity'] = similarities.diagonal()
    df.to_csv(output_csv, index=False)

    print(f"Cosine similarity calculated and saved to {output_csv}")
```

```python
input_csv = 'test questions with responses.csv'
output_csv = 'test questions final.csv'

calculate_cosine_similarity(input_csv, output_csv)
```

```python
import pandas as pd

def calculate_avg_similarity_all(file_path):
    df = pd.read_csv(file_path)
    avg_similarity = df['Cosine Similarity'].mean()
    return avg_similarity

def calculate_avg_similarity_non_null(file_path):
    df = pd.read_csv(file_path)
    non_null_df = df[df['Chatbot Response'].notnull()]
    print("length of not null rows:", len(non_null_df))
    avg_similarity_non_null = non_null_df['Cosine Similarity'].mean()
    return avg_similarity_non_null

avg_similarity_all = calculate_avg_similarity_all('test questions final.csv')
print(f"Average similarity (all rows): {avg_similarity_all}")

avg_similarity_non_null = calculate_avg_similarity_non_null('test questions final.csv')
print(f"Average similarity (non-null responses): {avg_similarity_non_null}")
```

# Gradio Interface

```python
import traceback

def generate_response_gradio(message, history):
  try:
    full_history = history + [{"role": "user", "content": message}]

    results = search(message, tokenizer, embedding_model, faiss_index, metadata_file="metadata.json", top_k=5)

    rag_context = retrieve(message, results)

    alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

    ### Instruction:
    {}

    ### Input:
    {}

    ### Response:
    {}"""

    instruction = """You're a mental health chatbot providing thoughtful, helpful, and safe responses to user queries.
    Your answers should be empathetic but realistic, focusing on providing actionable advice without being overly optimistic.
    You're going to get two inputs. One is the query from the patients, and another one is a RAG context retrieved from
    some mental health books. RAG context is supposed to help you frame answers. But sometimes, it might give you irrelavant information
    which is unrelated to the query. If that's the case, please ignore the RAG context and give appropriate answer to the query.
    In either answer the questions like a therapist."""

    prompt = f"Query: {message} RAG context: {rag_context}"

    inputs = tokenizer(
                        [
                            alpaca_prompt.format(
                              instruction,
                              prompt,
                              "",
                            )
                        ], return_tensors = "pt").to("cuda")

    outputs = model.generate(**inputs, max_new_tokens=1000, use_cache=True)
    response = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    clean_response= response[0].split("<|end_of_text|>")[0].split("### Response:\n")[-1].strip()
    full_history.append({"role": "assistant", "content": clean_response})

    return "", full_history

  except Exception as e:
    error_msg = f"An error occurred: {str(e)}\n{traceback.format_exc()}"
    print(error_msg)
    return f"I'm sorry, an error occurred: {str(e)}", history
```

```python
#gradio interface with history

with gr.Blocks(css=".gradio-container {background-color: #f0f0f0; font-family: Arial;}") as demo:
    gr.Markdown("<h1 style='text-align: center;'>Mental Health Chatbot</h1>")

    chatbot = gr.Chatbot(label="Chat History", type="messages")
    msg = gr.Textbox(placeholder="Type your message here...", label="Your Message")

    clear_button = gr.Button("Clear Chat")

    msg.submit(generate_response_gradio, [msg, chatbot], [msg, chatbot])

    clear_button.click(lambda: None, None, chatbot, queue=False)

demo.launch(debug= True)
```
